# PIL documentation: https://pillow.readthedocs.io/en/stable/

PIL stands for Python Imaging Library, which is a popular library in Python used for opening, manipulating, and saving many different image file formats.

```python
from PIL import Image, ImageEnhance
import numpy as np
```

```python
parrot = Image.open("parrot.png", "r")
```

```python
parrot
```

```python
parrot_array = np.asarray(parrot)
print(parrot_array)
```

```python
print(parrot_array.shape)
```

```python
parrot_bw = parrot.convert('L') # convert image to grayscale
parrot_bw
```

```python
parrot_bw_array = np.asarray(parrot_bw)
print(parrot_bw_array)
print(parrot_bw_array.shape)
```

```python
parrot_array[50:100, :]
```

```python
Image.fromarray(
    parrot_array[0:100, :] # Slice vertically, show all pixels horizontally
)
```

```python
Image.fromarray(
    # The first ":" means "Take all the pixels from the first dimension (height)"
    # The second ":" means: "Take all the pixels from the second dimension (width)"
    # The 0 means "Take only the red channel (the first color value - index 0)"
    # Try with 1 (green) or 2 (blue) instead and see what happens
    parrot_array[:, :, 0]
)
```

```python
color = np.full(shape=(250,250,3), fill_value=0)
color[:, :, 0] = 255
color = color.astype(np.uint8)

Image.fromarray(
        color
)
```

```python
def linear_combination(A, B, alpha):
    """
    Linearly combines A and B according to alpha.
    A and B are numpy arrays of the same shape.
    alpha is a scalar
    """
    lc = alpha * A + (1 - alpha) * B
    lc = lc.astype(np.uint8)
    return lc
```

```python
Image.fromarray(linear_combination(parrot_array, color, 0.5))
```

# OpenCV documentation: https://docs.opencv.org/4.x/d6/d00/tutorial_py_root.html

OpenCV stands for Open Source Computer Vision Library, and it is an open-source computer vision and machine learning software library. OpenCV was designed for computational efficiency and with a strong focus on real-time applications. It is written in C++ and has bindings for Python, Java, and other languages.

OpenCV includes several functions that are useful for pre-processing images and can make them suitable for processing by a CNN. It also includes some functions that allow you to use the deep learning models trained using different frameworks (like TensorFlow or PyTorch) directly in your OpenCV code. However, OpenCV itself is not a deep learning framework designed to create or train CNNs, but it can be used alongside deep learning libraries to implement and deploy CNNs for computer vision tasks.
