## API

- Una API (Interfaz de Programación de Aplicaciones) es una interfaz que permite acceder y recopilar datos de diversas fuentes de manera automatizada, facilitando la obtención de grandes volúmenes de información en ciencia de datos para su análisis.


- Ejemplos incluyen la API de Twitter, Google Maps, Spotify, o APIs financieras para datos de mercado.

```python
! pip install googlemaps
```

```python
import googlemaps
```

```python
# Reemplaza 'YOUR_API_KEY' con tu clave de API de Google Maps
API_KEY = ''
```

```python
# Inicializamos el cliente de Google Maps
gmaps = googlemaps.Client(key=API_KEY)
```

```python
def obtener_distancia_y_tiempo(origen, destino, modo="driving"):
    # Hacemos la solicitud a la API de Distance Matrix
    resultado = gmaps.distance_matrix(origins=origen, destinations=destino, mode=modo, units="metric")

    # Verificamos si la solicitud fue exitosa
    if resultado['status'] == 'OK':
        elemento = resultado['rows'][0]['elements'][0]
        if elemento['status'] == 'OK':
            distancia = elemento['distance']['text']
            duracion = elemento['duration']['text']
            return distancia, duracion
        else:
            return None, f"Error en el resultado: {elemento['status']}"
    else:
        return None, f"Error en la solicitud: {resultado['status']}"
```

```python
# Ejemplo de uso
origen = "Unicomfacauca"
destino = "Fundación Universitaria Konrad Lorenz"
modo = "driving"  # Opciones: driving (coche), walking (caminando), bicycling (bicicleta), transit (transporte público)
```

```python
distancia, duracion = obtener_distancia_y_tiempo(origen, destino, modo)
```

```python
print(f"Distancia: {distancia}")
print(f"Duración: {duracion}")
```

### Web scraping a una página de Wikipedia

```python
import requests
from bs4 import BeautifulSoup
```

```python
# URL de la página de Wikipedia que vamos a extraer
url = "https://es.wikipedia.org/wiki/Departamentos_de_Colombia"
```

```python
# Hacemos una solicitud HTTP para obtener el contenido de la página
response = requests.get(url)
```

```python
# Verificamos que la solicitud fue exitosa
if response.status_code == 200:
    # Parseamos el contenido HTML con BeautifulSoup
    soup = BeautifulSoup(response.content, 'html.parser')

else:
    print(f"Error al acceder a la página. Código de estado: {response.status_code}")
```

```python
soup
```

```python
# Encontramos el primer título de la página
title = soup.find('title')

# Mostramos el texto del título
print(f"Título de la página: {title.get_text()}")
```

```python
# Buscamos todos los encabezados de nivel 2 (que son las secciones principales)
section_titles = soup.find_all('h2')

print("Títulos de secciones:")
for title in section_titles:
  # Extraemos el texto de cada encabezado
  print(title.get_text(strip=True))
```

```python
# Encontramos todos los enlaces (etiquetas <a>)
links = soup.find_all('a')

# Imprimimos todos los enlaces
for link in links:
    # Obtenemos el texto del enlace y la URL
    link_text = link.get_text(strip=True)
    link_url = link.get('href')
    print(f"Texto: {link_text} - URL: {link_url}")
```

```python
# Encontramos un párrafo en la página
first_paragraph = soup.find('p')

# Obtenemos su nodo padre
parent = first_paragraph.parent

# Imprimimos el nodo padre y el contenido del párrafo
print(f"Nodo padre: {parent.name}")
print(f"Párrafo: {first_paragraph.get_text()}")

# También podemos navegar hacia el siguiente elemento en el mismo nivel
next_paragraph = first_paragraph.find_next_sibling('p')
if next_paragraph:
    print(f"Siguiente párrafo: {next_paragraph.get_text()}")
```

```python
# Encontrar todos los enlaces que empiezan con "https"
secure_links = soup.find_all('a', href=True)

# Filtramos solo los que contienen "https"
for link in secure_links:
    if 'https' in link['href']:
        print(f"Enlace seguro: {link['href']}")# Encontrar todas las tablas
tables = soup.find_all('table')

# Por cada tabla, encontramos todas las filas
for table in tables:
    rows = table.find_all('tr')

    # Extraemos el texto de cada celda en cada fila
    for row in rows:
        cells = row.find_all('td')
        cell_text = [cell.get_text(strip=True) for cell in cells]
        print(cell_text)
```

```python
# Extraemos todas las filas de la tabla
table_rows = table.find_all('tr')

# Extraemos los encabezados de la tabla
headers = [header.get_text(strip=True) for header in table_rows[0].find_all('th')]

# Extraemos los datos de cada fila
data = []
for row in table_rows[1:]:
    cells = row.find_all('td')
    data.append([cell.get_text(strip=True) for cell in cells])
```
