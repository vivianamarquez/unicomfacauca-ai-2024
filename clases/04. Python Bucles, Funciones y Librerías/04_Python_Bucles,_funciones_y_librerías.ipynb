## `For` (para)

Se utiliza para repetir código un **número concreto** de veces.


```python
for elemento in variable_iterable:
  # hacer algo con ese elemento
```

- Iterar es repetir un proceso (o un conjunto de instrucciones) varias veces
- La variable iterable puede ser una **lista**, una **cádena de texto**, o un **rango**.

````python
impares = [1,3,5,7,9,11,13]

for impar in impares:
  print(impar)
````

````python
impares = [1,3,5,7,9,11,13]

for impar in impares:
  print(impar+1)
````

````python
impares = [1,3,5,7,9,11,13]

for sanduche in impares:
  print(sanduche+1)
````

````python
impares = [1,3,5,7,9,11,13]

suma = 0

for impar in impares:
  suma = suma + impar

print(f"El total de la suma de esos impares es {suma}.")
````

````python
# Crear un arreglo de 380x380 píxeles con 3 canales de color (RGB), lleno de ceros (negro)
color = np.full(shape=(380,380,4), fill_value=0)

# Asignar el valor 255 al canal rojo (canal 0) para todos los píxeles
color[:, :, 0] = 255

# Asignar el valor 255 al canal alfa (canal 3) para hacer todos los píxeles completamente opacos
color[:, :, 3] = 255

# Convertir el arreglo al tipo de datos uint8 (valores entre 0 y 255, que son comunes para imágenes)
color = color.astype(np.uint8)

# Crear una imagen a partir del arreglo de colores y mostrarla
Image.fromarray(
        color
)
````

````python
def combinacion_lineal(A, B, alpha):
    """
    Combina linealmente A y B según el valor de alpha.
    A y B son arreglos de numpy con la misma forma.
    alpha es un escalar.
    """
    # Calcular la combinación lineal de A y B.
    # alpha * A da más peso a A, mientras que (1 - alpha) * B da más peso a B.
    combinacion = alpha * A + (1 - alpha) * B

    # Convertir el resultado al tipo de dato uint8 (valores entre 0 y 255)
    combinacion = combinacion.astype(np.uint8)

    # Devolver la combinación lineal final
    return combinacion
````

````python
Image.fromarray(combinacion_lineal(imagen_array, color, 0.5))
````
