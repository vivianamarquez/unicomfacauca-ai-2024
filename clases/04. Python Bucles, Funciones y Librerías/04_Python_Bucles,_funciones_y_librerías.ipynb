````python
def combinacion_lineal(A, B, alpha):
    """
    Combina linealmente A y B segÃºn el valor de alpha.
    A y B son arreglos de numpy con la misma forma.
    alpha es un escalar.
    """
    # Calcular la combinaciÃ³n lineal de A y B.
    # alpha * A da mÃ¡s peso a A, mientras que (1 - alpha) * B da mÃ¡s peso a B.
    combinacion = alpha * A + (1 - alpha) * B

    # Convertir el resultado al tipo de dato uint8 (valores entre 0 y 255)
    combinacion = combinacion.astype(np.uint8)

    # Devolver la combinaciÃ³n lineal final
    return combinacion
````
